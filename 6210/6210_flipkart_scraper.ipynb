## Flipkart Scraper

## Abstract-
This notebook deals with scraping data for books from the website flipkart. Flipkart is an Indian e-commerce website. We are going to scrape pricing information from this website on the basis of books.
We will be extracting Book Name, Number of Ratings, Ratings, Selling Price, Original Price, Language, Binding,
Author from the website.

```python
#Here we are importing all of our required python libraries 
from requests import get
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
#Here we are declaring all of our required list variables
book_name=[]     #Used to store the names of the books
no_of_ratings=[] #Used to store the number of user ratings on flipkart for that book
rating=[]        #Used to store the average rating given to the item by the users
selling_price=[] #Used to store the selling price of the books
original_price=[]#Used to store the Original price of the books
language=[]      #Used to store the language in which the book is written
binding=[]       #Used to store the binding type of the book
author=[]        #Used to store the author of the book
```

```python
#Variable used to store the link of the website
url = "https://www.flipkart.com/books/pr?sid=bks&q=books&otracker=categorytree&sort=popularity&page=1"
```

```python
counter = 1 #variable used to set the page counter
while counter != 51:  #beginning of the while loop, it will run 50 time to scrape 50 flipkart pages
    response = get(url)             #variable used to get the text from the url
    soup = bs(response.text, 'html.parser')     #variable used to get the Beautiful Soup objects
    library = soup.find(class_="_1HmYoV hCUpcT") #variable used to find the container block in flipkart
    book_container = library.find_all(class_ = "_3liAhj _1R0K0g") #variable use to get all books from the cuurent page


    for container in book_container:                    #parsing through everybook in the container

        name = container.find(class_="_2cLu-l").text    
        book_name.append(name)                           #appending the name of the author

        if container.find(class_="_38sUEc") is not None:
            no_of_ratings.append(container.find(class_="_38sUEc").text)
            no_of_ratings = [R.replace('(', '') for R in no_of_ratings]
            no_of_ratings = [R.replace(')', '') for R in no_of_ratings]
        else:
            no_of_ratings.append("N/A")                 #searching and cleaning the number of ratings
            
        if container.find(class_="hGSR34 _2beYZw") is not None:
            rating.append(container.find(class_="hGSR34 _2beYZw").text)
        else:
            rating.append("N/A")                         #searching and cleaning the ratings
    
        if container.find(class_="_1vC4OE") is not None:
            selling_price.append(container.find(class_="_1vC4OE").text)
        else:
            selling_price.append("N/A")                  #searching and cleaning the selling price
        
        if container.find(class_="_3auQ3N") is not None:
            original_price.append(container.find(class_="_3auQ3N").text)
        else:
            original_price.append(container.find(class_="_1vC4OE").text) #searching and cleaning the original price
        
        if container.find(class_="_1rcHFq") is not None:           #searching and cleaning the data for language, binding and author
            data_comb = container.find(class_="_1rcHFq").text
            data_list = data_comb.split(',')
            data_list = [x.strip(' ') for x in data_list]
            if len(data_list)==3:
                language.append(data_list[0])
                binding.append(data_list[1])
                author.append(data_list[2])
            elif  len(data_list)==2:
                language.append(data_list[0])
                binding.append(data_list[1])
                author.append("N/A")
            elif  len(data_list)==1:
                language.append(data_list[0])
                binding.append("N/A")
                author.append("N/A")
            elif  len(data_list)==4:
                language.append(data_list[0])
                binding.append(data_list[1])
                author.append(data_list[3]+data_list[2])
            elif  len(data_list)==4:
                language.append(data_list[0])
                binding.append(data_list[1])
                author.append(data_list[3]+data_list[2])
            else:
                language.append("N/A")
                binding.append("N/A")
                author.append("N/A")
        else:
            language.append("N/A")
            binding.append("N/A")
            author.append("N/A")
            
            
    counter = counter +1                    #incrementing the counter and concatinating it to the page 
    url = "https://www.flipkart.com/books/pr?sid=bks&q=books&otracker=categorytree&sort=popularity&page="+str(counter)
```

In the following set of code we deal with converting the individual lists into a dataframe

```python
flipkart_df = pd.DataFrame({"Book Name":book_name,
                        "Number Of Ratings":no_of_ratings,
                        "Rating":rating,
                        "Selling Price":selling_price,
                        "Original Price":original_price,
                        "Language":language,
                        "Binding":binding,
                        "Author":author,
                       })
```

```python
flipkart_df   #displaying the output of the data frame
```

```python
flipkart_df.to_csv("flipkart_prices.csv") #converting the dataframe into a csv file
```
